```python
import pandas as pd
df = pd.read_csv("titanic.csv")
df.head()
```

```python


inputs = df.drop('Survived',axis='columns')
target = df['Survived']
```

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['Pclass_n'] = le_company.fit_transform(inputs['Pclass'])
inputs['Sex_n'] = le_job.fit_transform(inputs['Sex'])
inputs['Age_n'] = le_degree.fit_transform(inputs['Age'])
inputs['Fare_n'] = le_company.fit_transform(inputs['Fare'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['Pclass','Sex','Age','Fare'],axis='columns')
```

```python
inputs_n
```

```python

target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n,target)
```
